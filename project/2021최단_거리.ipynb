```python
INF = 999
#             a    b     c     d     e     f     g     h     i 
adj_mat = [[  0,   9,   12,    8,    3,   13,   10,    7,   16 ], #a
           [  9,   0,    4,    6,    9,    5,   11,   12,   13 ], #b
           [ 12,   4,    0,    5,   12,    3,    7,   12,   10 ], #c
           [  8,   6,    5,    0,    8,    7,    3,   11,   11 ], #d
           [  3,   9,   12,    8,    0,   13,    8,    5,   13 ], #e
           [ 13,   5,    3,    7,   13,    0,    5,   10,    8 ], #f
           [ 10,  11,    7,    3,    8,    5,    0,    5,    6 ], #g
           [  7,  12,   12,   11,    5,   10,    5,    0,   10 ], #h
           [ 16,  13,   10,   11,   13,    8,    6,   10,   0 ], #i
           ]

node_num = len(adj_mat)
visited = [False] * node_num
print(visited)
distances = [INF] * node_num
print(distances)


"""
get_min_node 메소드
1. i가 0부터 node 갯수만큼 반복한다.
    i 노드가 방문한 적 없는 노드라면, 
        v에 i를 저장한다. 
        멈춘다. 
    
2. i가 0부터 node_num 만큼 반복한다.
        i가 방문하지 않았고, 방금 저장한 v보다 비용이 작은(기존의 신장트리로부터의 거리가 작은) 
        노드 i를 발견한다면
            v에 i를 덮어쓴다. 
    
    v를 리턴한다. 
"""


def get_min_node(node_num):
    print("get_min_node 함수 진입했다")
    for i in range(node_num):
        if not visited[i]:
            v = i
            break
    print("아직 방문하지 않은 v: ", v)
    for i in range(node_num):
        if not visited[i] and distances[i] < distances[v]: #방문안하고 무한대가 아닌 값이면 교체
            print("i: ", i, "\tv: ", v)
            print("distaces[i], distances[v]: ", distances[i], distances[v])
            v = i
            print("아직 방문 안됐고, 기존 v보다 distances값 작은 정점 발견 교체, 최종 v: ", v)

    return v


def prim(start, node_num):
    # distances 배열과 visted 배열 초기화
    for i in range(node_num):
        visited[i] = False
        distances[i] = INF

    # 시작노드의 distance 값을 0으로 초기화
    distances[start] = 0
    for i in range(node_num):
        print("--------------------")
        print("prim 메소드 내부 반복 시작, 반복 회차: ", i)
        node = get_min_node(node_num)
        visited[node] = True    # node 를 방문했다 표시
        print("\nget_min_node 에서 선택된 노드: ", node)

        for j in range(node_num):
            if adj_mat[node][j] != INF:
                if not visited[j] and adj_mat[node][j] < distances[j]:
                    print("방금 포함된 ", node, "와 연결되어있고, "
                                           "아직 방문하지 않았으며, "
                                           "기존의 신장트리로부터의 거리보다 작은 경로가 발견되어, "
                                           "distances[j]의 값을 갱신해야 하는 j: ", j)
                    distances[j] = adj_mat[node][j]

        print("distances: ", distances)
        print("--------------------")


print(prim(0, node_num))
print("distances: ", distances)
print("minimum cost: ", sum(distances))
```

그리디 알고리즘
지작지점에서 출발하여 가장 적은 weight만 따라서 가는 알고리즘

1. i에서 출발 
2. visited가 false인 i의 모든 경로를 탐색 가장 적은 가중치를 찾아서 visited를 true로 놓은다
3. 모든 visited가 전부 true인 경우 종료


```python
INF = 999
#             a    b     c     d     e     f     g     h     i 
adj_mat = [[INF,   9,   12,    8,    3,   13,   10,    7,   16 ], #a
           [  9, INF,    4,    6,    9,    5,   11,   12,   13 ], #b
           [ 12,   4,  INF,    5,   12,    3,    7,   12,   10 ], #c
           [  8,   6,    5,  INF,    8,    7,    3,   11,   11 ], #d
           [  3,   9,   12,    8,  INF,   13,    8,    5,   13 ], #e
           [ 13,   5,    3,    7,   13,  INF,    5,   10,    8 ], #f
           [ 10,  11,    7,    3,    8,    5,  INF,    5,    6 ], #g
           [  7,  12,   12,   11,    5,   10,    5,  INF,   10 ], #h
           [ 16,  13,   10,   11,   13,    8,    6,   10,  INF ], #i
           ]

node_num = len(adj_mat)
visited = [False] * node_num
weight = [INF] * node_num
loc=0
visited[0]=True


def min_weight(loc):    
    global v
    for i in range(node_num)    
        if visited and adj_mat[loc][i]<weight[loc]:
           weight[loc]=adj_mat[loc][i]
           v=i      
           i+=1
        else:
           i+=1       
    return weight[loc],v

def greedy(node_num):
    i=0
    while i < node_num:
        i_tem=i        
        visited[min_weight(i_tem)[1]]=True
        weight[min_weight(i_tem)[1]]=min_weight(i)[0]
        i_tem=min_weight(i_tem)[1]
        print('다음 번호는 %d'%(min_weight(i_tem)[1]))
        i+=1


    return visited,weight


print(greedy(node_num))
```

```python
import sys

input = sys.stdin.readline
n = 9
cities = [[  0,   9,   12,    8,    3,   13,   10,    7,   16 ], #a
           [  9,   0,    4,    6,    9,    5,   11,   12,   13 ], #b
           [ 12,   4,    0,    5,   12,    3,    7,   12,   10 ], #c
           [  8,   6,    5,    0,    8,    7,    3,   11,   11 ], #d
           [  3,   9,   12,    8,    0,   13,    8,    5,   13 ], #e
           [ 13,   5,    3,    7,   13,    0,    5,   10,    8 ], #f
           [ 10,  11,    7,    3,    8,    5,    0,    5,    6 ], #g
           [  7,  12,   12,   11,    5,   10,    5,    0,   10 ], #h
           [ 16,  13,   10,   11,   13,    8,    6,   10,   0 ], #i
           ]
VISITED_ALL = (1 << n) - 1

cache = [[None] * (1 << n) for _ in range(n)]
INF = float('inf')
idx = 1


def find_path(last, visited):
    if visited == VISITED_ALL:
        return cities[last][0] or INF
    if cache[last][visited] is not None:
        return cache[last][visited]

    tmp = INF
    for city in range(n):
        if visited & (1 << city) == 0 and cities[last][city] != 0:
            tmp = min(tmp, find_path(city, visited | (1 << city)) + cities[last][city])
    cache[last][visited] = tmp
    return tmp


print(find_path(0, 1 << 0))
```
